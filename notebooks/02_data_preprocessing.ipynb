# Satellite Land Cover Classification

## Notebook 02 — Feature Extraction

In this notebook we:

- Load the EuroSAT satellite images
- Extract handcrafted image features
- Create a structured machine learning dataset
- Save the processed dataset for model training

```python
from pathlib import Path

import pandas as pd
from PIL import Image
from tqdm.auto import tqdm

from landcover.feature_extraction import extract_features

from landcover.dataset import get_feature_names
```

```python
# ============================================================
# Dataset Paths
# ============================================================

RAW_DATASET = Path("../data/raw/eurosat_subset")

OUTPUT_DATASET = Path("../data/processed")

OUTPUT_DATASET.mkdir(
    parents=True,
    exist_ok=True
)

CSV_FILE = OUTPUT_DATASET / "satellite_landcover_features.csv"

print("Raw Dataset:")
print(RAW_DATASET)

print()

print("Output CSV:")
print(CSV_FILE)
```

```python
# ============================================================
# Find Land Cover Classes
# ============================================================

classes = sorted([

    folder.name

    for folder in RAW_DATASET.iterdir()

    if folder.is_dir()

])

print("Classes Found:")

classes
```

```python
# ============================================================
# Extract Features
# ============================================================

rows = []

for landcover_class in classes:

    image_files = sorted(

        (RAW_DATASET / landcover_class).glob("*.jpg")

    )

    print(f"{landcover_class}: {len(image_files)} images")

    for image_path in tqdm(

        image_files,

        desc=landcover_class

    ):

        image = Image.open(image_path).convert("RGB")

        features = extract_features(image)

        features["Label"] = landcover_class

        rows.append(features)
```

```python
# ============================================================
# Build Dataset
# ============================================================

df = pd.DataFrame(rows)

print("Dataset Shape:")

print(df.shape)

print()

display(df.head())
```

```python
# ============================================================
# Save Dataset
# ============================================================

df.to_csv(

    CSV_FILE,

    index=False

)

print("Dataset saved successfully!")

print(CSV_FILE)
```

```python
# ============================================================
# Verify Saved Dataset
# ============================================================

verification_df = pd.read_csv(CSV_FILE)

print("Verification Successful!")

print()

print(verification_df.shape)

display(verification_df.head())
```

```python
# ============================================================
# Feature Extraction Summary
# ============================================================

print("=" * 60)
print("FEATURE EXTRACTION COMPLETE")
print("=" * 60)

print(f"Images Processed : {len(df):,}")

print(f"Classes          : {df['Label'].nunique()}")

print(f"Features         : {len(get_feature_names())}")

print()

print("Feature Columns")

for feature in get_feature_names():

    print(f"  • {feature}")

print()

print("Class Distribution")

print(df["Label"].value_counts())

print()

print("Dataset Saved To")

print(CSV_FILE)

print()

print("=" * 60)
```
