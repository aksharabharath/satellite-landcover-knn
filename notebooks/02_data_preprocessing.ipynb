# Satellite Land Cover Classification
## 02 - Feature Extraction and Data Preprocessing

```python
from pathlib import Path

import numpy as np
import pandas as pd
from PIL import Image
```

```python
# Path to our dataset
dataset_path = Path("../data/raw/eurosat_subset")

# The three land cover classes we're using
classes = ["Forest", "River", "Residential"]

print(dataset_path)
print(classes)
```

```python
# Count how many images we have in each class

for land_cover in classes:
    
    class_folder = dataset_path / land_cover
    
    image_files = list(class_folder.glob("*.jpg"))
    
    print(f"{land_cover}: {len(image_files)} images")
```

```python
# This list will store information about every image
dataset = []
```

```python
for land_cover in classes:

    class_folder = dataset_path / land_cover

    image_files = list(class_folder.glob("*.jpg"))

    print(f"\nProcessing {land_cover}...")

    for image_path in image_files[:3]:

        image = Image.open(image_path)
        image_array = np.array(image)

        mean_red = image_array[:, :, 0].mean()
        mean_green = image_array[:, :, 1].mean()
        mean_blue = image_array[:, :, 2].mean()

        dataset.append({
            "Mean_R": mean_red,
            "Mean_G": mean_green,
            "Mean_B": mean_blue,
            "Label": land_cover
        })
```

```python
# Show the first 5 entries in our dataset
dataset[:5]
```

```python
# Convert our list into a Pandas DataFrame
df = pd.DataFrame(dataset)

# Display the first few rows
df.head()
```

```python
# Start fresh
dataset = []

for land_cover in classes:

    class_folder = dataset_path / land_cover

    image_files = list(class_folder.glob("*.jpg"))

    print(f"Processing {land_cover}...")

    for image_path in image_files:

        # Open image
        image = Image.open(image_path)

        # Convert to NumPy array
        image_array = np.array(image)

        # Mean RGB
        mean_red = image_array[:, :, 0].mean()
        mean_green = image_array[:, :, 1].mean()
        mean_blue = image_array[:, :, 2].mean()

        # Standard deviation RGB
        std_red = image_array[:, :, 0].std()
        std_green = image_array[:, :, 1].std()
        std_blue = image_array[:, :, 2].std()

        # Store one row
        dataset.append({
            "Mean_R": mean_red,
            "Mean_G": mean_green,
            "Mean_B": mean_blue,
            "Std_R": std_red,
            "Std_G": std_green,
            "Std_B": std_blue,
            "Label": land_cover
        })

print(f"\nFinished! Processed {len(dataset)} images.")
```

```python
# Convert the dataset into a DataFrame
df = pd.DataFrame(dataset)

# Display the first 5 rows
df.head()
```

```python
# Create a new feature: average brightness
df["Brightness"] = (df["Mean_R"] + df["Mean_G"] + df["Mean_B"]) / 3

# Check the updated DataFrame
df.head()
```

```python
# Save the processed dataset
df.to_csv("../data/processed/satellite_landcover_features.csv", index=False)

print("Dataset saved successfully!")
```
